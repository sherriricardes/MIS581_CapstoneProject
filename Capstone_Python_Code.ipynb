```python
import pandas as pd
from functools import reduce

# Load individual NHANES files
bmi = pd.read_sas("BMI.xpt")[['SEQN', 'BMXBMI']]
cholesterol = pd.read_sas("Cholesterol.xpt")[['SEQN', 'LBXTR', 'LBDLDL']]
crp = pd.read_sas("CReactiveProtein.xpt")[['SEQN', 'LB2CRP']]
demographics = pd.read_sas("Demographics.xpt")[['SEQN', 'RIAGENDR', 'DMDHREDU', 'RIDAGEYR']]
DNAmethylation = pd.read_sas("DNAmethylation.sas7bdat", format='sas7bdat', encoding='utf-8')[
    ['SEQN', 'HorvathAge', 'HannumAge', 'PhenoAge', 'GrimAgeMort',
     'SkinBloodAge', 'DunedinPoAm', 'ZhangAge', 'VidalBraloAge',
     'WeidnerAge', 'LinAge']]
fitness = pd.read_sas("Fitness.xpt")[['SEQN', 'CVDFITLV']]
glucose = pd.read_sas("Glucose.xpt")[['SEQN', 'LB2GLU', 'LB2IN']]
health = pd.read_sas("HealthStatus.xpt")[['SEQN', 'HSD010']]
activity = pd.read_sas("PhysicalActivity.xpt")[['SEQN', 'PAQ180']]
telomere = pd.read_sas("Telomere.xpt")[['SEQN', 'TELOMEAN']]
vitaminD = pd.read_sas("VitaminD.xpt")[['SEQN', 'LBDVIDMS']]
weight = pd.read_sas("WeightHistory.xpt")[['SEQN', 'WHD020']]

# Rename columns
bmi.rename(columns={'BMXBMI': 'BMI'}, inplace=True)
cholesterol.rename(columns={'LBXTR': 'Triglycerides', 'LBDLDL': 'LDL'}, inplace=True)
crp.rename(columns={'LB2CRP': 'CReactiveProtein'}, inplace=True)
demographics.rename(columns={'DMDHREDU': 'Education', 'RIDAGEYR': 'Age'}, inplace=True)
glucose.rename(columns={'LB2GLU': 'Glucose', 'LB2IN': 'Insulin'}, inplace=True)
telomere.rename(columns={'TELOMEAN': 'TelomereLength'}, inplace=True)
vitaminD.rename(columns={'LBDVIDMS': 'VitaminD'}, inplace=True)
weight.rename(columns={'WHD020': 'Weight'}, inplace=True)

# Merge all datasets using SEQN
dfs_to_merge = [bmi, cholesterol, crp, demographics, DNAmethylation, fitness,
                glucose, health, activity, telomere, vitaminD, weight]
merged_df = reduce(lambda left, right: pd.merge(left, right, on='SEQN', how='outer'), dfs_to_merge)

# Map categorical variables to readable text
merged_df['Gender'] = merged_df['RIAGENDR'].map({1: 'Male', 2: 'Female'})
merged_df['GeneralHealth'] = merged_df['HSD010'].map({
    1: 'Excellent', 2: 'Very good', 3: 'Good', 4: 'Fair', 5: 'Poor',
    7: 'Refused', 9: "Don't know"
})
merged_df['PhysicalActivityLevel'] = merged_df['PAQ180'].map({
    1: 'Sit most of the day',
    2: 'Stand or walk a lot',
    3: 'Lift light loads or climb often',
    4: 'Do heavy work or lift heavy loads',
    7: 'Refused', 9: "Don’t know"
})
merged_df['FitnessLevel'] = merged_df['CVDFITLV'].map({1: 'Low', 2: 'Moderate', 3: 'High'})

# Drop original coded columns
merged_df.drop(columns=['RIAGENDR', 'HSD010', 'PAQ180', 'CVDFITLV'], inplace=True)

# Preview and save
print("Merged dataset preview:")
print(merged_df.head())
print(f"Final dataset shape: {merged_df.shape}")

# Save to file
merged_df.to_csv("BioAge_Dataset_Updated1.csv", index=False)
```

```python
import pandas as pd
import statsmodels.api as sm
from sklearn.impute import SimpleImputer
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Load dataset
df = pd.read_csv("BioAge_Dataset_Updated.csv").copy()

# Define features (X) and target (y)
X = df[['BMI', 'Triglycerides', 'LDL', 'CReactiveProtein', 'Age',
        'Glucose', 'Insulin', 'TelomereLength', 'VitaminD', 'Weight',
        'PhysicalActivityLevel']].copy()
y = df['GrimAgeMort'].copy()

# Convert categorical variables
X['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})
X = pd.concat([X, pd.get_dummies(df['GeneralHealth'], prefix='GeneralHealth', drop_first=True)], axis=1)
X = pd.concat([X, pd.get_dummies(df['FitnessLevel'], prefix='FitnessLevel', drop_first=True)], axis=1)

# Drop FitnessLevel_Low and FitnessLevel_Moderate dummy variables
X = X.drop(columns=[col for col in X.columns if 'FitnessLevel_Low' in col or 'FitnessLevel_Moderate' in col], errors='ignore')

# Impute missing values
imputer = SimpleImputer(strategy='mean')
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

# Drop rows where y is missing
y_notna = y.notna()
X_imputed = X_imputed[y_notna]
y = y[y_notna]

# Add constant
X_with_const = sm.add_constant(X_imputed)

# Calculate VIF
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X_with_const.values, i) for i in range(X_with_const.shape[1])]
vif["features"] = X_with_const.columns

# Print VIF
print("Variance Inflation Factors:")
print(vif)

# Fit oridinary least squares model
model = sm.OLS(y, X_with_const).fit()

# Print full summary
print("\nModel Summary:")
print(model.summary())
```

```python
# Analysis for Each Research Question

import pandas as pd
import statsmodels.api as sm
from sklearn.impute import SimpleImputer
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Load dataset
df = pd.read_csv("BioAge_Dataset_Updated.csv").copy()

# Target variable (GrimAgeMort - estimated biological age)
y = df['GrimAgeMort'].copy()

# Define biomarker-only features for Research Question 1
X_biomarkers = df[['TelomereLength', 'Triglycerides', 'LDL', 'Glucose', 'Insulin']].copy()

# Define features for Research Question 2 (biomarkers + lifestyle)
X_biomarkers_lifestyle = df[['aTelomereLength', 'Triglycerides', 'LDL', 'Glucose', 'Insulin',
                             'PhysicalActivityLevel']].copy()

# Define full model features for Research Question 3 (adding demographics + health indicators)
X_full = df[['TelomereLength', 'Triglycerides', 'LDL', 'Glucose', 'Insulin',
             'PhysicalActivityLevel', 'CReactiveProtein', 'VitaminD', 'Weight', 'BMI', 'Age']].copy()

# Map gender for RQ2 and RQ3
df['Gender_numeric'] = df['Gender'].map({'Male': 0, 'Female': 1})

# Add GeneralHealth and FitnessLevel dummies for RQ3
general_health_dummies = pd.get_dummies(df['GeneralHealth'], prefix='GeneralHealth', drop_first=True)
fitness_level_dummies = pd.get_dummies(df['FitnessLevel'], prefix='FitnessLevel', drop_first=True)

# Add to X_full for RQ3
X_full = pd.concat([X_full, df[['Gender_numeric']], general_health_dummies, fitness_level_dummies], axis=1)

# Drop problematic dummy columns
X_full = X_full.drop(columns=[col for col in X_full.columns if 'FitnessLevel_Low' in col or 'FitnessLevel_Moderate' in col], errors='ignore')

# Function to preprocess and model
def run_ols(X, y, description="Model"):
    # Impute missing values
    imputer = SimpleImputer(strategy='mean')
    X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)
    
    # Drop missing y rows
    y_notna = y.notna()
    X_imputed = X_imputed[y_notna]
    y_clean = y[y_notna]
    
    # Add constant
    X_with_const = sm.add_constant(X_imputed)
    
    # Calculate VIF
    vif = pd.DataFrame()
    vif["VIF Factor"] = [variance_inflation_factor(X_with_const.values, i) for i in range(X_with_const.shape[1])]
    vif["features"] = X_with_const.columns
    print(f"\nVariance Inflation Factors for {description}:")
    print(vif)
    
    # Fit OLS model
    model = sm.OLS(y_clean, X_with_const).fit()
    
    # Print model summary
    print(f"\nModel Summary for {description}:")
    print(model.summary())

# Run models for each Research Question

# Research Question 1: Biomarkers only
run_ols(X_biomarkers, y, "Research Question 1: Biomarkers Only")

# Research Question 2: Biomarkers + Lifestyle
run_ols(X_biomarkers_lifestyle, y, "Research Question 2: Biomarkers + Lifestyle")

# Research Question 3: Full Model with Inflammation, Demographics, Lifestyle
run_ols(X_full, y, "Research Question 3: Full Model with Inflammation and Health Indicators")
```

```python
import matplotlib.pyplot as plt

# Get fitted values (predicted y) and residuals (errors)
fitted_values = model.fittedvalues
residuals = model.resid

# Create the plot
plt.figure(figsize=(8,5))
plt.scatter(fitted_values, residuals, alpha=0.5)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Predicted Values (Fitted)')
plt.ylabel('Residuals')
plt.title('Residuals vs. Predicted Values')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

def run_random_forest(X, y, description="Random Forest Model"):
    # Drop missing target values
    y_clean = y.dropna()
    X_clean = X.loc[y_clean.index]

    # Impute missing predictor values
    imputer = SimpleImputer(strategy='mean')
    X_imputed = pd.DataFrame(imputer.fit_transform(X_clean), columns=X.columns, index=X_clean.index)

    # Fit Random Forest Regressor
    rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
    rf_model.fit(X_imputed, y_clean)

    # Predict and evaluate
    y_pred = rf_model.predict(X_imputed)
    r2 = r2_score(y_clean, y_pred)
    rmse = np.sqrt(mean_squared_error(y_clean, y_pred))
    mae = mean_absolute_error(y_clean, y_pred)

    print(f"\n{description} Results:")
    print(f"R² Score: {r2:.4f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
    print(f"Mean Absolute Error (MAE): {mae:.4f}")

    # Feature importance
    feature_importances = pd.Series(rf_model.feature_importances_, index=X.columns)
    print("\nTop 10 Important Features:")
    print(feature_importances.sort_values(ascending=False).head(10))

run_random_forest(X_full, y, "Research Question 3: Random Forest Model")
```
